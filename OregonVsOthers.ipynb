## <font color="blue">Data comparison between Oregon, US, Canada, France, England, and Germany </font>

Importing in the basic libraries, and then reading in the data from our Github repository

```python
import pandas as pd
import matplotlib.pyplot as plt

comp_data = pd.read_csv('https://raw.githubusercontent.com/AxisMeetsWorld/Oregon_National_Europe_college/main/OregonComparison.csv')
comp_data
```

<font color='purple'>**Now let's get two dataframes for the tuition and for the cost of living and then stack these up on a bar chart.**

```python

comp_tuition=comp_data[comp_data['Expense']=='Tuition']
comp_living=comp_data[comp_data['Expense']!='Tuition']
comp_living
```

```python
Oregon_Tot = comp_tuition[comp_tuition['Location']=='Oregon']['Cost in US Dollars'] + comp_living[comp_living['Location']=='Oregon']['Cost in US Dollars']

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

ax.bar(x=comp_tuition['Location'], height=comp_tuition['Cost in US Dollars'] ,width =0.45, color='aqua')
ax.bar(x=comp_living['Location'], height=comp_living['Cost in US Dollars'], bottom = comp_tuition['Cost in US Dollars'], width=0.45,color='b')
ax.set_title("Oregon Resident In and Out of State")
ax.legend(labels=['Tuition','Cost of living'])
plt.show()
```
